```python
import pandas as pd
```

```python
df = pd.read_csv('/opt/data/in_rolls/in_rolls_state_name_age_sex_fn_ln.csv.gz', usecols=['state', 'birth_year', 'last_name', 'sex'])
df
```

```python
# take out the data from Andhra
df.drop(df[df.state=='andhra'].index, inplace=True)
df
df
```

```python
# drop birth year before 1900
df.drop(df[df.birth_year < 1900].index, inplace=True)
```

```python
# Drop NaN and a bunch of weird names for now
df.dropna(subset=['last_name'], inplace=True)
df.drop(df[~df.last_name.str.contains('^[a-z]{2,}')].index, inplace=True)
df.drop(df[df.last_name.str.contains('\.')].index, inplace=True)
df
```

```python
gdf = df.groupby(['birth_year', 'last_name']).sex.value_counts().unstack(fill_value=0).reset_index()
gdf.columns = ['birth_year', 'last_name', 'n_female', 'n_male', 'n_third_gender']
gdf['sex_ratio'] = gdf['n_male'] / gdf['n_female']
gdf
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
from labellines import labelLine, labelLines
```

```python
# We ignore the long tail of last names because we cannot learn sex ratios reliably where we don't have a large enough n.
# Instead, we focus on last names shared by at least 50,000 people.
ldf = gdf.groupby(['last_name']).agg({'n_female': 'sum', 'n_male': 'sum'}).reset_index()
ldf = ldf[ldf.n_male + ldf.n_female > 50000]

pdf = gdf[gdf.last_name.isin(ldf.last_name)].groupby(['birth_year']).agg({'n_female': 'sum', 'n_male': 'sum'}).reset_index()
pdf = pdf[pdf.n_male + pdf.n_female > 1000].copy()
pdf['sex_ratio'] = pdf.n_male / pdf.n_female
pdf.set_index('birth_year', inplace=True)
pdf
```

```python
pdf.sex_ratio.plot(figsize=(16, 12))
```

```python
# graphs that show within last name, over time, where we have enough data (> 1000 per year)
pdf = gdf[gdf.last_name.isin(ldf.last_name)].copy()
pdf = gdf[gdf.n_male + gdf.n_female > 1000].copy()
pdf
```

```python
# filter range of sex_ratio
pdf = pdf[(pdf.sex_ratio > 0.5) & (pdf.sex_ratio < 5)]
```

```python
adf = pdf.groupby(['birth_year', 'last_name']).agg({'sex_ratio': 'mean'}).unstack()['sex_ratio']
adf
```

```python
# only keep last names where we have data for at least 30 years
bdf = adf.T[adf.T.count(axis=1) >= 30]
bdf
```

```python
bdf.T.plot(figsize=(16, 12))
labelLines(plt.gca().get_lines(), xvals=(1975, 1990), zorder=5)
```
