### Check amount of male/female group by year for 'rao' and 'singh'

```python
rao_df = df[df.last_name=='rao'].groupby(['birth_year']).sex.value_counts().unstack(fill_value=0)
rao_df
```

```python
singh_df = df[df.last_name=='singh'].groupby(['birth_year']).sex.value_counts().unstack(fill_value=0)
singh_df
```
